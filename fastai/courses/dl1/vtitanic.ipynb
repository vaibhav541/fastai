```python
y=t['Survived'].astype('float32')
drop_elements = ['PassengerId', 'Name', 'Ticket', 'Cabin','Survived']
t = t.drop(drop_elements, axis = 1)
```

```python
t['Embarked'] = t['Embarked'].fillna('2')
t['Age'] = t['Age'].fillna(t['Age'].mean()).astype('float32')
t=t.astype('float32')
```

```python
t.describe()
```

```python
label_csv=f'{PATH}train.csv'
n=len(list(open(label_csv)))-1
val_idxs=get_cv_idxs(n)
```

```python
cat_vars=['Sex','Embarked']
y=y.astype('float32')
```

```python
md = ColumnarModelData.from_data_frame(PATH, val_idxs,t,y ,cat_flds=cat_vars, bs=128,)
```

```python
emb_szs=[(2,2),(3,2)]
```

```python
m = md.get_learner(emb_szs, len(t.columns)-len(cat_vars),
                   0.04, 1, [1000,500], [0.001,0.01])
lr = 1e-3
```

```python
def inv_y(a): return np.exp(a)
def exp_rmspe(y_pred, targ):
    targ = inv_y(targ)
    pct_var = (targ - inv_y(y_pred))/targ
    return math.sqrt((pct_var**2).mean())
```

```python
m.fit(lr, 200)
```

```python
x,y=m.predict_with_targs()
```
