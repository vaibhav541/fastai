```python
from PIL import Image
import numpy as np
import pandas as pd
import os
```

```python
tc=f'data/hwd/train.csv'
df=pd.read_csv(tc)
```

```python
images = df.iloc[:,1:].values
images = images.astype(np.float)

# For labels
label = df['label']
```

```python
images.shape
```

```python
for i in range(0, len(images)):

    #read the correct label
    correct_label = label[i]

    #split the data into training and validation sets
    if np.random.rand() < 0.8:
        folder = 'train/'
    else:
        folder = 'valid/'
    
    img = images[i][:]

    #reshape into 28x28 pic
    img = img.reshape(28,28)

    #we need three channels into the picture
    img = np.stack((img,)*3,axis = -1)

    #change the data type to int8
    img = np.uint8(img)

    #create PIL Image
    new_img = Image.fromarray(img)

    #save the .jpg into correct folder
    new_img.save(f'data/hwd/' + folder + str(correct_label) + '/' + str(i) + '.jpg', 'JPEG')
```
